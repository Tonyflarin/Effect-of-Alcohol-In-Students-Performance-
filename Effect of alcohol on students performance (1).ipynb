```python
#Marks of male students
men=n[n["Your Sex?"]=="Male"]
plt.figure(figsize=(15,5))
plt.hist(men["Your Matric (grade 12) Average/ GPA (in %)"],color="black",bins=8)
plt.xlabel("marks")
plt.ylabel("male")
plt.show()
```

```python
#Marks of female students
female=n[n["Your Sex?"]=="Female"]
plt.figure(figsize=(15,5))
plt.hist(female["Your Matric (grade 12) Average/ GPA (in %)"],color="black")
plt.xlabel("marks")
plt.ylabel("female")
plt.show()
```
